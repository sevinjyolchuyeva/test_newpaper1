```python
from gensim.models import FastText
import numpy as np


path1='training_input.txt'
path2='training_output.txt'
path3='test_input.txt'
path4='test_output.txt'
path5='dev_input.txt' # validation data
path6='dev_output.txt'

l,l1,l2,l3=[],[],[],[]

with open(path1,'r') as f:
    for line in f:
        line=line.strip()
        l.append([line])
        l1.append([line])
        
with open(path3,'r') as f:
    for line in f:
        line=line.strip()
        l.append([line])
        l2.append([line])
        
with open(path5,'r') as f:
    for line in f:
        line=line.strip()
        l.append([line])
        l3.append([line])

```

```python
model = FastText(l,window=10,hs=1,min_n=1,alpha=0.25,word_ngrams=3, max_n=6,min_count=1,iter=50)


```

```python
Xtrain=np.zeros((len(l1),100))
for i in range(len(l1)):
    wo=l1[i]
    Xtrain[i]=model[wo]

print(Xtrain.shape)

Xdev=np.zeros((len(l2),100))
for i in range(len(l2)):
    wo=l2[i]
    Xdev[i]=model[wo]
    
Xtest=np.zeros((len(l3),100))
for i in range(len(l3)):
    wo=l3[i]
    Xtest[i]=model[wo]

```

```python
l1,l2,l3,l=[],[],[],[]

with open(path2,'r') as f:
    for line in f:
        line=line.strip()
        ll=line.split(' ')
        l+=ll
        l1.append(ll)

with open(path4,'r') as f:
    for line in f:
        line=line.strip()
        ll=line.split(' ')
        l+=ll
        l2.append(ll)
        
with open(path6,'r') as f:
    for line in f:
        line=line.strip()
        ll=line.split(' ')
        l+=ll
        l3.append(ll)

        
chars_o= sorted(list(set(l)))
len_out=len(chars_o)
print('out:',len_out)
char_indices_o = dict((c, i) for i, c in enumerate(chars_o))
ndices_char_o= {v:k for k, v in char_indices_o.items()}

maxlen=23
print ('length_all_phoneme: ',len_out)
print ('all_phoneme: ',chars_o)                           

print('Vectorization of output...')  

Ytrain = np.zeros((len(l1), maxlen, len_out), dtype=np.bool)
for i, sentence in enumerate(l1):
        for t, char in enumerate(sentence):
            Ytrain[i, t, char_indices_o[char]] = 1

print('shape_of_output_train: ' , Ytrain.shape)



Ydev = np.zeros((len(l2), maxlen, len_out), dtype=np.bool)
for i, sentence in enumerate(l2):
        for t, char in enumerate(sentence):
            Ydev[i, t, char_indices_o[char]] = 1

print('shape_of_output_dev: ' , Ydev.shape)


Ytest = np.zeros((len(l3), maxlen, len_out), dtype=np.bool)
for i, sentence in enumerate(l3):
        for t, char in enumerate(sentence):
            Ytest[i, t, char_indices_o[char]] = 1

print('shape_of_output_test: ' , Ytest.shape)
```

```python
np.savez('training1.npz',x=Xtrain,y=Ytrain)
np.savez('dev1.npz',x=Xdev,y=Ydev)
np.savez('test1.npz',x=Xtest,y=Ytest)
```
